```python
import ROOT as R
import sys
```

```python
tc = R.TCanvas()
my_hist = R.TH1F("my_hist", "My First Histogram", 100, -5, 5)
my_hist.FillRandom("gaus", 10000)
my_hist.Draw()
tc.Draw()
```

Here is an example of a more complex canvas.  <br>  Very often you will run a C++ program to do some heavy processing and write the results (histograms, etc) to a file (eg TFile).  Then it is very convenient to use a notebook like this to access your output data file, read in the results and work on the plotting.  We can easily manipulate the plots and re-run the cell below to adjust how things look.

Have a look at the notebook called SpeedTest to see an example of the speed difference for executing a simple function to do an equivalent calculation in C++ and Python.

```python
#alternatively, we can make our TCanvas and subdivide it if we want
tc=R.TCanvas("tc","My Canvas")
tc.Divide(2,2)   # columns, rows  2x3 TPads in this case
tc.cd(1)         # TPads are numbered going left to right, top to bottom
my_hist.Draw()
tc.cd(2)
my_hist.Draw("E")           # draw with error bars
tc.cd(3)
my_hist.Draw("E")           # draw with error bars
h2=R.TH1F(my_hist)
h2.SetLineColor(R.kRed)
h2.SetLineWidth(2)
h2.Draw("L,same")      # overlay with lines through the points
tc.cd(4)
my_hist.Draw("HBAR")        # draw as horizontal bar chart
tc.Draw()
```

```python
tc.Print("first.pdf")       # save our canvas as a PDF
```
